```python
import pandas as pd
import numpy as np
import datetime as dt
```

# Get data from http://www.nyc.gov/html/tlc/html/about/trip_record_data.shtml
Data collection (2018 year 1-6 months) and cleaning

Data Cleaning: check every file that it contains the respected month data, if such data entries exists delete those entries


```python
i=0
while i<6:
    url = 'https://s3.amazonaws.com/nyc-tlc/trip+data/yellow_tripdata_2018-0'+str(i+1)+'.csv'
    df=pd.read_csv(url)
    df['tpep_pickup_datetime'] = df.tpep_pickup_datetime.apply(lambda x:dt.datetime.strptime(x,"%Y-%m-%d %H:%M:%S"))
    df['tpep_dropoff_datetime'] = df.tpep_dropoff_datetime.apply(lambda x:dt.datetime.strptime(x,"%Y-%m-%d %H:%M:%S"))
    df = df[df['tpep_pickup_datetime'].dt.month == i+1]
    df = df[df['tpep_pickup_datetime'].dt.year == 2018]
    df = df[df['tpep_dropoff_datetime'].dt.year == 2018]
    if i==0:
        print(url)
        resultdata=df
    else:
        print(url)
        resultdata=resultdata.append(df, ignore_index=True)
    i+=1
resultdata.head()
```

```python
resultdata=resultdata.dropna()
resultdata=resultdata.reset_index(drop=True)
```

# Please write a Python program that calculates the average trip length of all Yellow Taxis for a month.

```python
resultdata.groupby([pd.to_datetime(resultdata.tpep_pickup_datetime).dt.strftime('%Y-%m')])['trip_distance'].mean().reset_index(name='Monthly Average')
```

# Please write a Python program that calculates the average trip length of all Yellow Taxis for a Day

```python
result=resultdata.groupby([pd.to_datetime(resultdata.tpep_pickup_datetime).dt.strftime('%Y-%m-%d')])['trip_distance'].mean().reset_index(name='Monthly Average')
result
```

# Extend this to a data pipeline that can ingest new data and calculates the 30 day rolling average trip length.

```python
MovingAverage = pd.rolling_mean(result,31)
MovingAverage
```
